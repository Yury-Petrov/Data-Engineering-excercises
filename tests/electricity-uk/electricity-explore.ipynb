```python
from pyspark.shell import spark
from pyspark.sql.functions import col, lit,split, expr
```

```python
spark

df = spark.read.option('header', True).csv('BANES Energy Data Electricity.csv')
```

```python
time_intervals = '+'.join(list(filter(lambda c: '_' in c, df.columns)))
time_intervals
```

```python
df = df.withColumn('Country', lit('United Kingdom'))
df = df.where(col('postcode').isNotNull())

postcode_split = split(col('postcode'), ' ')
df = df.withColumn('postcode', postcode_split.getItem(0))
```

```python
df.withColumn('total_consumption', expr(time_intervals))\
    .coalesce(1)\
    .write.format('csv').options(header=True).mode('overwrite')\
    .save('./country')
```
